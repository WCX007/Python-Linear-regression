```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('boston.csv').to_numpy()
```

```python
n,m=data.shape
```

```python
b = np.ones((n,1))
data = np.column_stack((b,data))
```

```python
data.shape
```

```python
train = data[:406,:]
test = data[406:,:]
```

```python
x_train = train[:,:14]
y_train = train[:,-1].reshape(len(y_train),1)
```

```python
x_test = test[:,:14]
y_test = test[:,-1]
```

```python
w = np.linalg.inv(x_train.T.dot(x_train)).dot(x_train.T).dot(y_train)
```

```python
print(w)
```

```python
pre = x_test.dot(w)
```

```python
for i in range(len(y_test)):
    print(y_test[i],pre[i][0])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(x_train,y_train)
```

```python
p = reg.predict(x_test)
```

```python
for i in range(len(y_test)):
    print(y_test[i],p[i][0])
```
